```python
data = pd.read_csv('data.csv', encoding = 'utf-8')
data
```

```python
VNs = []
for i in data.index:
    new_VN = data['VN'][i]
    VNs.append(ViTokenizer.tokenize(new_VN).replace('\u200b', ''))
VNs
```

```python
data['VN'] = VNs
```

```python
data.to_csv('data.csv', encoding = 'utf-8', index = False)
```
